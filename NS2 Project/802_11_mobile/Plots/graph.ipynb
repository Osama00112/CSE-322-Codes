```python
df1 = pd.read_csv('Flows1.csv')
df2 = pd.read_csv('mFlows1.csv')
fig, ax = plt.subplots()
ax.plot(df1['Flows'], df1['Throughput'], label='Original', color = "blue")
ax.plot(df2['Flows'], df2['Throughput'], label='Modified', color = "green")

ax.scatter(df1['Flows'], df1['Throughput'], color = "red")
ax.scatter(df2['Flows'], df2['Throughput'], color = "orange")

ax.set_ylabel("Throughput (kbits/sec)")
ax.set_xlabel("Flows")
ax.set_title("Flows vs Throughput")
ax.legend()
ax.grid(True)
#plt.ylim(ymin = 0)
plt.xlim(xmin = 10)
ax.set_xticks([10,20,30,40,50])
plt.tight_layout()
fig.savefig("graphs/Flows1.png");
plt.show()
plt.close(fig)
```
