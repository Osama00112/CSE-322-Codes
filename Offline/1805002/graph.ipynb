```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Area1.csv')

print(df.to_string())
ax = df.plot.scatter(x='Area', y='Throughput')
df.plot.line(x='Area', y='Throughput', ax=ax)
ax.set_xticks([250,500,750,1000,1250])
plt.grid()
plt.ylim(ymin = 0)
plt.xlim(xmin = 250)
plt.show()
```
